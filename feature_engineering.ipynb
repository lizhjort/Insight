```
checkouts = pd.read_pickle('/content/gdrive/My Drive/insight_data/checkouts.pkl.gz')
```

```
merged_inv_check = pd.merge(inventory, checkouts, on=['BibNumber'], how='left')
```

```
merged_inv_check = merged_inv_check.drop(['CheckoutDateTime', 'Publisher',	'ReportDate',	'ItemCount',	'ItemBarcode',	'CallNumber'], axis=1)
```

```
merged_inv_check.to_pickle('/content/gdrive/My Drive/insight_data/merged_inv_check.pkl.gz')
```

```
checkouts['CheckoutDate'] = checkouts['CheckoutDateTime'].dt.normalize()
```

```
checkouts_grouped = checkouts.groupby(by=['BibNumber', 'CheckoutDate'], as_index=False).count()
```

```
checkouts_grouped = checkouts_grouped.drop(['CallNumber', 'ItemTitle', 'CheckoutDateTime'], axis=1)
```

```
checkouts_grouped = checkouts_grouped.rename(index=str, columns={"ItemBarcode": "CheckoutCount"})
```

```
checkouts_grouped.sort_values(by='CheckoutDate')
```

```
checkouts_grouped.shape
```

```
checkouts_grouped['CheckoutCount'].describe()
```

```
unique_checkouts = checkouts_grouped['BibNumber'].unique()
unique_checkouts = pd.DataFrame(data=unique_checkouts)
unique_checkouts.shape
```

```
years = ['1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']
inventory = inventory[inventory['PublicationYear'].str.contains('|'.join(years), na=False)]
```

```
inventory.head()
```

```
merged_inv = pd.merge(inventory, checkouts_grouped, on=['BibNumber'], how='outer')
```

```
merged_inv = merged_inv.sort_values(by='CheckoutDate')
```

```
merged_inv.shape
```

```
merged_inv = merged_inv.drop(['ReportDate', 'ItemCount'], axis=1)
```

```
merged_inv.head(15)
```

```
merged = merged_inv.groupby([pd.Grouper(key='CheckoutDate', freq='W'), 'BibNumber'])['CheckoutCount'].sum()
```

```
merged = merged.reset_index()
```

```
merged
```

```
unique_date = merged['CheckoutDate'].unique()
unique_date = pd.DataFrame(data=unique_date)
unique_date.shape
```
